#**8.** *파이썬에서 time()을 호출하면 1970년 1월 1일 이후 흘러온 전체 초가 반환됩니다. 이것을 적절하게 나누거나 나머지를 구해서 지금 현재의 시와 분을 출력해보세요.(실제 현재 시각이 나와야 합니다.)*
8.PNG

```python
import time

seconds = int(time.time())
minutes = (seconds // 60) % 60
hours = (seconds // 3600) % 24
print("현재 시간" , (hours, minutes, seconds))
```
